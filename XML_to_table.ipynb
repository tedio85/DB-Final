```python
'''parse XML roadlevel_info.xml'''

import xml.etree.ElementTree as ET
from psycopg2 import sql

tree = ET.parse('./data/roadlevel_info.xml')
root = tree.getroot()
```

```python
'''form sql statements'''

infos = root.findall('Infos/Info')

for info in infos:
    
    data = (
     info.get('locationpath'),
     info.get('speedlimit'),
     info.get('fromkm'),
     info.get('tokm'),
     info.get('roadtype'),
     info.get('startlocationpoint'),
     info.get('endlocationpoint'),
     info.get('roadsection'),
     info.get('sourceid'),
     info.get('routeid'))
    
    cursor.execute("insert into Highway values (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s);", data)
        
```

```python
'''show the result of the table Highway'''

cursor.execute("""SELECT * FROM Highway;""")
rows = cursor.fetchall()
print(rows)
```
